```python
import json
import os


results_dir = "webvoyager/20250420_171609"

# Load the failed tasks from the JSONL file
failed_tasks_path = os.path.join(results_dir, "failed_tasks.jsonl")
failed_tasks = []

with open(failed_tasks_path, "r") as f:
    for line in f:
        failed_tasks.append(json.loads(line))

failed_task_ids = [task["id"] for task in failed_tasks]
print(failed_task_ids)
```

```python
def inspect_failed_task(task_id):
    metadata_path = os.path.join(results_dir, f"{task_id}/metadata.json")
    with open(metadata_path, "r") as f:
        metadata = json.load(f)

    task = metadata["objective"]
    print(f"Task: {task}")
    iterations = metadata["iterations"]
    print(f"Tterations: {iterations}")

    eval_reasoning = metadata["auto_eval"]["reasoning"]
    print(f"Evaluation Reasoning:\n{eval_reasoning}")

    message_history = metadata["message_history"]
    print(message_history)


# for task_id in failed_task_ids:
#     inspect_failed_task(task_id)
#     print("-" * 100)
inspect_failed_task("Apple--20")
```
